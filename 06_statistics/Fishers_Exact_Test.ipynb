```python
import pandas as pd
from scipy.stats import fisher_exact
```

File reading

```python
all_lineage = dir + "Fungi-RNA-221013-completed_lineage_table_simplified_final.csv"
all_lineage = pd.read_csv(all_lineage)
all_lineage
```

```python
virus_hit_lineage = dir + "Fungi-RNA-221013-completed_lineage_table_simplified_final_virushunter_hits.csv"
virus_hit_lineage = pd.read_csv(virus_hit_lineage)
virus_hit_lineage
```

# Ascomycota vs Basidiomycota

```python
all_lineage_count = all_lineage['phylum'].value_counts().rename_axis('phylum').reset_index(name='count')
all_lineage_count
```

```python
virus_hit_lineage_count = virus_hit_lineage['phylum'].value_counts().rename_axis('phylum').reset_index(name='count')
virus_hit_lineage_count
```

```python
asco_T = all_lineage_count.loc[all_lineage_count['phylum'] == 'Ascomycota', 'count'].iloc[0]
basi_T = all_lineage_count.loc[all_lineage_count['phylum'] == 'Basidiomycota', 'count'].iloc[0]
asco_virusP = virus_hit_lineage_count.loc[virus_hit_lineage_count['phylum'] == 'Ascomycota', 'count'].iloc[0]
basi_virusP = virus_hit_lineage_count.loc[virus_hit_lineage_count['phylum'] == 'Basidiomycota', 'count'].iloc[0]
asco_virusN = asco_T - asco_virusP 
basi_virusN = basi_T - basi_virusP

# generate dataframe for fisher's exact test
df_fisher_1 = pd.DataFrame({
    'Ascomycota': [asco_virusP, asco_virusN],
    'Basidiomycota': [basi_virusP, basi_virusN]
    }, index=['virus_positive', 'virus_negative'])

df_fisher_1
```

```python
odd_ratio, p_value = fisher_exact(df_fisher_1, alternative='two-sided')
print(p_value)
print(odd_ratio)
```

```python
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(df_fisher_1)
print(p)
```
